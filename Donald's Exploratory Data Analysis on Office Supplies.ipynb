```python
### Merging the east_ and west_profit on index to create east_west_profit using an inner join
## East = E ,West = W , Central = C , South = S
## Profit = Prof
east_west_profit = east_profit.merge(west_profit, left_index = True, right_index = True , how = 'inner')
east_west_profit 
```

```python
### merging east_west_profit with central profit to create east_west_central_profit using an inner join
east_west_central_profit = east_west_profit.merge(central_profit, left_index = True, right_index = True, how = 'inner')
east_west_central_profit
```

```python
### Creatimg the all_region_profit by merging east_west_central_profit with the south_profit using an inner join
all_region_profit = east_west_central_profit.merge(south_profit, left_index = True, right_index = True , how = 'inner')
all_region_profit
```
